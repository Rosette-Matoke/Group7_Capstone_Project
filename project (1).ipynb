# üöÄ PROJECT SUMMARY
This project developed an intelligent medical chatbot system designed to improve patient guidance and streamline the disease diagnosis process. ## 6.3 Symptom Co-occurrence Network

```python
#!pip install networkx
```

```python
import networkx as nx
# Drop the disease column
symptom_matrix = df.drop(columns=['diseases'])

# Count frequency of each symptom 
symptom_freq = symptom_matrix.sum().sort_values(ascending=False)

# Select top 50 most common symptoms
top_symptoms = symptom_freq.head(20).index

# Filter to only those 50
symptom_matrix_reduced = symptom_matrix[top_symptoms]

#  Build co-occurrence matrix 
co_matrix = symptom_matrix_reduced.T.dot(symptom_matrix_reduced)

# Create network graph 
G = nx.from_pandas_adjacency(co_matrix)

#  Filter to strongest relationships
edges = [(u, v) for u, v, d in G.edges(data=True) if d['weight'] > 50]  # adjust threshold as needed
H = G.edge_subgraph(edges)

# Visualize 
plt.figure(figsize=(12, 10))
nx.draw(
    H,
    with_labels=True,
    node_color='lightgreen',
    edge_color='lightblue',
    node_size=800,
    font_size=8
)
plt.title('Top 20 Symptom Co-occurrence Network', fontsize=14, fontweight='bold')
plt.show()
```

#### Clusters or Groups:

You can see distinct clusters this are groups of symptoms that co-occur more frequently with each other than with others.

For example **fever** ,**cough** ,**headache** ,**vomiting** and **nausea** cluster likely represents infectious or flu-like illnesses.

**Skin_swelling** **skin_lesion** **rash** **abnormal_appearing_skin** cluster likely represents dermatological or allergic conditions.

**Sharp_chest** **pain‚Äìshortness_of_breath** and **dizziness** might relate to cardiopulmonary issues.


**Highly Connected Symptoms (Hubs):**

Symptoms like fever, cough, and headache are in the center ‚Äî they co-occur with many other symptoms.

These act as hub nodes, indicating they are common across multiple disease types.

**Peripheral Nodes:**

Symptoms on the outer edges (like depressive_or_psychotic_symptoms or skin_swelling) have fewer connections showing they appear in more specific or isolated conditions.

NB :
Symptoms that appear isolated in a co-occurrence network‚Äîlike **vaginal bleeding after menopause**or **poor circulation** typically indicate that they do not frequently co-occur with other symptoms in the dataset. This often means they are diagnostically specific, context-dependent, or underreported.

### 6.3.1  Centrality Analysis
This helps identify the most influential symptoms in your network.

üí° Interpretation:
- Degree Centrality: Symptoms that co-occur with many others.
- Betweenness Centrality: Bridge symptoms that connect clusters.
- Closeness Centrality: Symptoms that are close to all others and are good for early detection.

```python
#Compute centrality scores
degree_centrality = nx.degree_centrality(H)
betweenness_centrality = nx.betweenness_centrality(H)
closeness_centrality = nx.closeness_centrality(H)

# Sort and display top 5 symptoms by each centrality
print("Top 5 by Degree Centrality:")
print(sorted(degree_centrality.items(), key=lambda x: x[1], reverse=True)[:5])

print("\nTop 5 by Betweenness Centrality:")
print(sorted(betweenness_centrality.items(), key=lambda x: x[1], reverse=True)[:5])

print("\nTop 5 by Closeness Centrality:")
print(sorted(closeness_centrality.items(), key=lambda x: x[1], reverse=True)[:5])
```

 **1. Headache: The Dominant Symptom**
 
- Top in all three metrics: degree, betweenness, and closeness.
- Highly connected it Co-occurs with many other symptoms.
- Bridge symptom it links different clusters, possibly appearing in diverse conditions
- Central to network quickly reachable from other symptoms, making it a key candidate for early screening .
  
 **2. Sharp Abdominal Pain**
  
- Strong across all metrics, second only to headache.
- Likely part of multiple syndromes could be gastrointestinal, gynecological or even referred pain.
- Its high betweenness suggests it may connect unrelated symptom groups, making it diagnostically significant.
  
 **3. Vomiting, Dizziness, Cough**
  
- High degree and closeness, but lower betweenness.
- These symptoms are common companions in many conditions (e.g., flu, food poisoning, vertigo).
- They‚Äôre not bridges, but they‚Äôre central within their own clusters useful for narrowing down diagnoses once a cluster is identified.
- 
 **4. Back Pain & Abnormal Appearing Skin**
  
- High betweenness but not top in degree or closeness.
- These symptoms may act as gateways between clusters‚Äîless frequent but strategically important.

**Plotting feature importance for logistic regression**


```python
def plot_logreg_feature_importance(model, feature_names, top_n=20):
    """
    Plot overall (aggregate) feature importance for multiclass Logistic Regression.

    The importance is computed as the mean absolute coefficient across all classes.
    """
    coef = model.coef_

    # Take absolute value and average across classes
    mean_importance = np.mean(np.abs(coef), axis=0)

    # Create DataFrame
    importance_df = pd.DataFrame({
        'Feature': feature_names,
        'Importance': mean_importance
    }).sort_values(by='Importance', ascending=False)

    # Select top features
    top_features = importance_df.head(top_n)

    # Plot
    plt.figure(figsize=(10, 8))
    plt.barh(top_features['Feature'], top_features['Importance'], color='cyan')
    plt.gca().invert_yaxis()
    plt.title("Overall Logistic Regression Feature Importance")
    plt.xlabel("Mean Absolute Coefficient")
    plt.ylabel("Feature")
    plt.tight_layout()
    plt.show()
```

```python
def plot_rf_feature_importance(model, feature_names, top_n=20):
    """
    Plot top N feature importances for a Random Forest model.
    """

    importances = model.feature_importances_
    indices = np.argsort(importances)[::-1][:top_n]
    top_features = np.array(feature_names)[indices]
    top_importances = importances[indices]

    plt.figure(figsize=(10, 6))
    plt.barh(range(len(top_features)), top_importances[::-1])
    plt.yticks(range(len(top_features)), top_features[::-1])
    plt.xlabel("Feature Importance")
    plt.title(f"Top {top_n} Important Features (Random Forest)")
    plt.show()
```

```python
symptom_columns = df_model.columns[1:]  # all columns except 'diseases'

def combine_symptoms(row):
    symptoms = [col.replace("_", " ") for col in symptom_columns if row[col] == 1]
    return ", ".join(symptoms)

df_model["symptoms_text"] = df_model.apply(combine_symptoms, axis=1)

df_model = df_model[["symptoms_text", "diseases"]]
df_model.to_csv("my_data.csv", index=False)
print(df_model.head())

```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import accuracy_score, classification_report
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, LSTM, Dense, Dropout, Bidirectional
from tensorflow.keras.callbacks import EarlyStopping
import joblib
```

```python
#Load data
df = pd.read_csv("my_data.csv")  # columns: symptoms_text, diseases

X = df["symptoms_text"].astype(str)
y = df["diseases"].astype(str)

# Encode labels
label_encoder = LabelEncoder()
y_encoded = label_encoder.fit_transform(y)

# Train-test split

X_train, X_test, y_train, y_test = train_test_split(
    X, y_encoded, test_size=0.2, random_state=42, stratify=y_encoded
)

#Tokenization & Padding
MAX_WORDS = 5000
MAX_LEN = 100

tokenizer = Tokenizer(num_words=MAX_WORDS, oov_token="<OOV>")
tokenizer.fit_on_texts(X_train)

X_train_seq = tokenizer.texts_to_sequences(X_train)
X_test_seq = tokenizer.texts_to_sequences(X_test)

X_train_pad = pad_sequences(X_train_seq, maxlen=MAX_LEN, padding="post", truncating="post")
X_test_pad = pad_sequences(X_test_seq, maxlen=MAX_LEN, padding="post", truncating="post")
```

```python
# Neural Network (Lightweight)
model = Sequential([
    Embedding(input_dim=MAX_WORDS, output_dim=128, input_length=MAX_LEN),
    Bidirectional(LSTM(64, dropout=0.3, recurrent_dropout=0.2, return_sequences=False)),
    Dense(64, activation='relu'),
    Dropout(0.3),
    Dense(len(np.unique(y_encoded)), activation='softmax')
])

model.compile(loss='sparse_categorical_crossentropy', optimizer='adam', metrics=['accuracy'])

#Train model
early_stop = EarlyStopping(monitor='val_loss', patience=3, restore_best_weights=True)

history = model.fit(
    X_train_pad, y_train,
    validation_split=0.2,
    epochs=10,
    batch_size=32,
    callbacks=[early_stop],
    verbose=1
)
```

```python
 Evaluate performance
train_loss, train_acc = model.evaluate(X_train_pad, y_train, verbose=0)
test_loss, test_acc = model.evaluate(X_test_pad, y_test, verbose=0)

print(f"Training Accuracy: {train_acc:.4f}")
print(f"Test Accuracy: {test_acc:.4f}")
print("\nClassification Report:")
y_pred = np.argmax(model.predict(X_test_pad), axis=1)
print(classification_report(y_test, y_pred, target_names=label_encoder.classes_))
```

```python
# Get predictions and probabilities
y_pred_proba = model.predict(X_test_pad)
y_pred = np.argmax(y_pred_proba, axis=1)

# Compute other metrics
accuracy = accuracy_score(y_test, y_pred)
macro_precision = precision_score(y_test, y_pred, average='macro')
macro_recall = recall_score(y_test, y_pred, average='macro')
macro_f1 = f1_score(y_test, y_pred, average='macro')
logloss = log_loss(y_test, y_pred_proba)

# Display overall performance summary
print("\nOverall Neural Network (Text Classification) Performance:")
print(f"Accuracy       : {accuracy:.4f}")
print(f"Macro Precision: {macro_precision:.4f}")
print(f"Macro Recall   : {macro_recall:.4f}")
print(f"Macro F1-score : {macro_f1:.4f}")
print(f"Log Loss       : {logloss:.4f}")
```

### 8.1 HOW DO WE DEFINE SUCCESS? (METRICS)
our model is a multi-class medical diagnosis classifier, predicting a disease based on symptoms.
In this kind of task, accuracy alone is not enough, because some classes (diseases) might have more samples than others and false negatives can be more dangerous than false positives.

**Healthcare & diagnosis require reliability over raw accuracy**

A model that‚Äôs 90% accurate but miss diagnoses 10% of real patients (low recall) is dangerous.
Hence, *recall (catching all true cases)* is a priority.

**Precision** ensures responsible predictions
You don‚Äôt want to label many healthy people as ‚Äúinfected‚Äù or ‚Äúcritical‚Äù.High precision ensures your model only flags a condition when it‚Äôs confident.

**F1-score** gives a fair balance
Especially if diseases are not equally represented, F1 helps ensure performance is consistent across all classes.

**Log Loss** validates trust
In a diagnostic setting, the confidence (probability) of prediction matters (for example, ‚Äú90% malaria‚Äù vs ‚Äú55% malaria‚Äù).
Low log loss means the model‚Äôs probability predictions are trustworthy and suitable for decision support.


**NB:** In our medical diagnosis model, the most important metrics are Recall, F1-score, Precision, and Log Loss.
Recall ensures that no true disease cases are missed, Precision ensures responsible predictions, F1-score maintains balance between both, and Log Loss validates the model‚Äôs confidence.
These metrics will drive our decision making when choosing a model to deploy.

```python
def medical_chatbot(user_input):
    import random
    greetings = ["hello", "hi", "hey", "good morning", "good evening"]
    responses = ["Hello! How can I assist you today?", "Hi there! What symptoms are you experiencing?"]
    
    if any(g in user_input.lower() for g in greetings):
        return random.choice(responses)
    
    seq = tokenizer.texts_to_sequences([user_input])
    pad = pad_sequences(seq, maxlen=MAX_LEN, padding='post')
    pred = model.predict(pad)
    disease = label_encoder.inverse_transform([np.argmax(pred)])[0]
    return f"Based on your symptoms, you might have: {disease}"


#Try the chatbot
print("\nChatbot ready! Type 'exit' to stop.")
while True:
    user_input = input("You: ")
    if user_input.lower() == "exit":
        print("Chatbot: Goodbye!")
        break
    print("Chatbot:", medical_chatbot(user_input))
```
