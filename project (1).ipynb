# üëî  1. BUSINESS UNDERSTANDING
In the healthcare sector, patient‚Äìdoctor communication plays a crucial role in ensuring timely diagnosis, effective treatment, and overall patient satisfaction. However, doctors often face high workloads and limited consultation time, while patients struggle to access reliable medical advice promptly. This imbalance between healthcare demand and availability creates a significant gap in early diagnosis and patient support.

To bridge this gap, the development of a medical chatbot offers a practical and innovative solution. The chatbot functions as a virtual health assistant, capable of providing patients with instant, preliminary health guidance, symptom analysis, and responses to common medical questions. By doing so, it enhances accessibility to healthcare information and supports both patients and professionals in improving communication, efficiency, and care outcomes.

The introduction of a medical chatbot comes with several advantages. It provides 24/7 accessibility, ensuring that patients can receive guidance at any time, even outside normal clinic hours. It also helps reduce the workload for healthcare professionals by handling repetitive and low-complexity interactions, allowing doctors to focus on more critical cases. Additionally, the chatbot contributes to faster preliminary diagnoses by collecting patient symptoms ahead of consultation, and it improves cost efficiency by reducing the need for constant human involvement in basic queries. Moreover, it can collect valuable data on patient interactions for use in research and healthcare system improvements. Importantly, chatbots deliver consistent and standardized information, ensuring accuracy and eliminating human variability in responses.

Despite these benefits, several limitations must be acknowledged. Chatbots possess limited diagnostic accuracy, as they cannot fully understand a patient‚Äôs medical history or complex symptom combinations. They also lack emotional intelligence, meaning they cannot replicate the empathy and reassurance that human healthcare providers offer. Data privacy and security risks pose additional challenges since medical data is highly sensitive and must be protected from unauthorized access. Furthermore, internet dependence can limit accessibility in low-connectivity areas, and regulatory concerns arise from the need to ensure compliance with medical and ethical standards. Lastly, the chatbot may struggle with language nuances or contextual understanding, particularly in diverse linguistic environments.

To mitigate these challenges, a range of strategies will be implemented. The chatbot will combine rule-based and machine learning techniques to balance accuracy with control, and it will clearly communicate that its responses are for informational purposes only. Predefined empathetic response patterns will be incorporated to enhance user experience, while data encryption and anonymization protocols will safeguard user privacy. To improve accessibility, an offline or SMS-based version may be developed for users in low-bandwidth regions. Compliance with medical and data protection regulations such as HIPAA and GDPR will be ensured through close collaboration with medical experts. Finally, the chatbot will be trained on diverse and localized text datasets to improve language comprehension and contextual relevance.

Through these measures, the medical chatbot will provide an accessible, cost-effective, and secure digital healthcare assistant that enhances patient engagement while maintaining ethical, safe, and reliable healthcare delivery.

## 6.3 Symptom Co-occurrence Network

```python
#!pip install networkx
```

```python
import networkx as nx

# Compute co-occurrence matrix
symptom_matrix = df.drop(columns=['diseases'])
co_matrix = symptom_matrix.T.dot(symptom_matrix)

# Build network graph
G = nx.from_pandas_adjacency(co_matrix)

# Filter to strongest relationships
edges = [(u,v) for u,v,d in G.edges(data=True) if d['weight'] > 50]  # adjust threshold
H = G.edge_subgraph(edges)

plt.figure(figsize=(12,10))
nx.draw(H, with_labels=True, node_color='lightblue', edge_color='grey', node_size=800, font_size=8)
plt.title('Symptom Co-occurrence Network')
plt.show()
```

**What Clustered Symptoms Reveal**

When symptoms are tightly connected in your network‚Äîlike **bloating**, **abdominal pain**, **nausea**, **vomiting**, and **diarrhea** they form a symptom cluster. This means:
- They frequently co-occur in patient records.
- They likely share a common etiology (cause or disease mechanism).
- They may represent a syndrome or a stage of disease progression.

Symptoms that appear isolated in a co-occurrence network‚Äîlike **vaginal bleeding after menopause**or **poor circulation** typically indicate that they do not frequently co-occur with other symptoms in the dataset. This often means they are diagnostically specific, context-dependent, or underreported.

 These clusters are powerful tools for diagnosis, treatment planning, and predictive modeling.

### 6.3.1  Centrality Analysis
This helps identify the most influential symptoms in your network.

üí° Interpretation:
- Degree Centrality: Symptoms that co-occur with many others.
- Betweenness Centrality: Bridge symptoms that connect clusters.
- Closeness Centrality: Symptoms that are close to all others and are good for early detection.

```python
#Compute centrality scores
degree_centrality = nx.degree_centrality(H)
betweenness_centrality = nx.betweenness_centrality(H)
closeness_centrality = nx.closeness_centrality(H)

# Sort and display top 5 symptoms by each centrality
print("Top 5 by Degree Centrality:")
print(sorted(degree_centrality.items(), key=lambda x: x[1], reverse=True)[:5])

print("\nTop 5 by Betweenness Centrality:")
print(sorted(betweenness_centrality.items(), key=lambda x: x[1], reverse=True)[:5])

print("\nTop 5 by Closeness Centrality:")
print(sorted(closeness_centrality.items(), key=lambda x: x[1], reverse=True)[:5])
```

 **1. Headache: The Dominant Symptom**
 
- Top in all three metrics: degree, betweenness, and closeness.
- Highly connected it Co-occurs with many other symptoms.
- Bridge symptom it links different clusters, possibly appearing in diverse conditions
- Central to network quickly reachable from other symptoms, making it a key candidate for early screening .
  
 **2. Sharp Abdominal Pain**
  
- Strong across all metrics, second only to headache.
- Likely part of multiple syndromes could be gastrointestinal, gynecological or even referred pain.
- Its high betweenness suggests it may connect unrelated symptom groups, making it diagnostically significant.
  
 **3. Vomiting, Dizziness, Cough**
  
- High degree and closeness, but lower betweenness.
- These symptoms are common companions in many conditions (e.g., flu, food poisoning, vertigo).
- They‚Äôre not bridges, but they‚Äôre central within their own clusters useful for narrowing down diagnoses once a cluster is identified.
- 
 **4. Back Pain & Abnormal Appearing Skin**
  
- High betweenness but not top in degree or closeness.
- These symptoms may act as gateways between clusters‚Äîless frequent but strategically important.

This heatmap shows how strongly specific symptoms are associated with different diseases‚Äîdarker cells indicate stronger correlations, helping identify diagnostic patterns and symptom clusters.

- Headache and Concussion (0.52): A classic symptom of head trauma.

- Skin swelling and Dental caries (0.48): May reflect oral infections spreading to surrounding tissue.
- leg pain and injury to leg (0.5) : show the relation between getting injured and pain.

```python
def plot_detached_pie(Categories, top_n=5):
    """
    Plots a detached (exploded) pie chart for the top N disease categories.
    """
    
    if 'Disease_Count' not in Categories.columns:
        Categories['Disease_Count'] = Categories['Diseases'].apply(lambda x: len(str(x).split(',')))
    
    
    top_df = Categories.sort_values(by='Disease_Count', ascending=False).head(top_n)
    
    # Detach each slice slightly
    explode = [0.05] * len(top_df)
    
    # Plot pie chart
    plt.figure(figsize=(8,8))
    plt.pie(
        top_df['Disease_Count'],
        labels=top_df['Category'],
        autopct='%1.1f%%',
        startangle=140,
        colors=plt.cm.Paired.colors,
        explode=explode,
        shadow=True
    )
    plt.title(f"Top {top_n} Disease Categories (Detached Pie Chart)", fontsize=14, fontweight='bold')
    plt.show()


plot_detached_pie(Categories, top_n=5)
```

```python
#### Random grid search CV for Neural Network Classifier
from sklearn.model_selection import RandomizedSearchCV
from sklearn.neural_network import MLPClassifier
# Create a fresh MLPClassifier
nn_base = MLPClassifier(random_state=42)
# Define hyperparameter grid
param_dist = {
    'hidden_layer_sizes': [(50,), (100,), (150,)],
    'activation': ['tanh', 'relu'],
    'solver': ['sgd', 'adam'],
    'alpha': [0.0001, 0.001, 0.01],
    'learning_rate': ['constant', 'adaptive'],
}
# Run randomized search
random_search_nn = RandomizedSearchCV(
    estimator=nn_base,
    param_distributions=param_dist,
    n_iter=20,
    cv=3,
    verbose=1,
    random_state=42,
    n_jobs=1  # Use single job to avoid parallel processing issues
)
# Fit the model
random_search_nn.fit(X_train, y_train)
print("Best Hyperparameters for Neural Network:", random_search_nn.best_params_)
# Get best model
best_nn_model = random_search_nn.best_estimator_
#### Evaluate best model on training set
y_train_pred_best_nn = best_nn_model.predict(X_train)
accuracy_train_best_nn = accuracy_score(y_train, y_train_pred_best_nn)
print(f"Best Neural Network Model Training Accuracy: {accuracy_train_best_nn:.4f}")
# Evaluate best model on test set
y_pred_best_nn = best_nn_model.predict(X_test)
accuracy_best_nn = accuracy_score(y_test, y_pred_best_nn)
print(f"Best Neural Network Model Test Accuracy: {accuracy_best_nn:.4f}")
#### Evaluate best model on validation set
y_val_pred_best_nn = best_nn_model.predict(X_val)
accuracy_val_best_nn = accuracy_score(y_val, y_val_pred_best_nn)
print(f"Best Neural Network Model Validation Accuracy: {accuracy_val_best_nn:.4f}")
```
