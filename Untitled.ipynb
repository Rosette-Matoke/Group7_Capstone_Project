```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
import numpy as np
# Example: Patient symptoms as text
X = df['Patient_text']
y = df['Doctor_text']   # or diagnosis, category, etc.

X_cleaned = X.fillna('').astype(str)

# 1️⃣ Vectorize the input text
tfidf = TfidfVectorizer(max_features=5000)
X_tfidf = tfidf.fit_transform(X_cleaned)

# 2️⃣ Encode the target labels
le = LabelEncoder()
y_encoded = le.fit_transform(y)

# 3️⃣ Split and train
X_train, X_test, y_train, y_test = train_test_split(X_tfidf, y_encoded, test_size=0.4, random_state=0)

model = MultinomialNB()
model.fit(X_train, y_train)

# 4️⃣ Evaluate
accuracy = model.score(X_test, y_test)
print("Model accuracy:", accuracy)
```
